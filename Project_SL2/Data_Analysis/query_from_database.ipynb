```python
# Empty dataframe

df = pd.DataFrame(columns=['ID', 'User', 'Location', 'Motion', 'Humidity (%)', \
    'Temperature (F)', 'Time'])
```

```python
# SQl query

query = "SELECT * FROM SensorData;"
cursor.execute(query)
data = cursor.fetchall()

for row in data:
    if row[0] <= 1469:
        continue
    df.loc[df.shape[0]+1] = row
    
```

```python
# Data cleaning

df['Humidity (%)'] = df['Humidity (%)'].apply(lambda x: x[:-1] if type(x) == str else x).astype(float)
df['Temperature (F)'] = df['Temperature (F)'].apply(lambda x: x[:-2] if isinstance(x, str) else x).astype(float)
```

```python
df['Motion'] = df['Motion'].replace({'0': False, '1': True})
```
